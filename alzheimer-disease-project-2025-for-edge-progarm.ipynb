## 5. Correlation Analysis
Check which features are most correlated with the `Diagnosis`.

```python
# Compute correlation matrix
correlation_matrix = data.corr(numeric_only=True)
diagnosis_correlation = correlation_matrix['Diagnosis'].sort_values(ascending=False)

# Show top 10 correlations
diagnosis_correlation
```

```python
# Plot top 10 positively/negatively correlated features with Diagnosis
top_features = diagnosis_correlation[1:11]  # exclude Diagnosis itself
plt.figure(figsize=(10,6))
sns.barplot(x=top_features.values, y=top_features.index, palette='coolwarm')
plt.title('Top 10 Features Correlated with Diagnosis')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.grid(True)
plt.show()
```


### 📊 Interpretation of Bar Plot
- This bar plot shows the top 10 features most correlated with Alzheimer's diagnosis.
- **MemoryComplaints** and **BehavioralProblems** have the highest positive correlations, suggesting they are strong predictors.
- Lifestyle-related features like **CholesterolHDL** and **BMI** show weaker correlations.
- This validates clinical understanding that cognitive and behavioral symptoms are key indicators.


## 6. Full Correlation Heatmap

```python
plt.figure(figsize=(14,8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Features')
plt.show()
```


### 🔥 Interpretation of Full Correlation Heatmap
- The full heatmap shows how all features relate to each other.
- Features such as **MMSE**, **FunctionalAssessment**, and **ADL** are **negatively correlated** with Diagnosis.
- Positive correlation among symptom-related features and Diagnosis is observed.
- Identifies groups of related features and potential multicollinearity.


## 7. Focused Heatmap on Top Correlated Features

```python
top_features_list = diagnosis_correlation.index[1:11].tolist()
top_features_list.append('Diagnosis')
focused_corr_matrix = data[top_features_list].corr()

plt.figure(figsize=(10,8))
sns.heatmap(focused_corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Focused Correlation Heatmap (Top Features Related to Diagnosis)')
plt.show()
```


### 🎯 Interpretation of Focused Heatmap
- This zooms into the most influential features for Diagnosis.
- **MemoryComplaints**, **BehavioralProblems**, and **DifficultyCompletingTasks** cluster together.
- **MMSE**, **ADL**, and **FunctionalAssessment** form a negatively correlated group.
- Highlights redundancy among similar features and guides feature selection.

### 🚀 Model Training – Using Top Correlated Features




```python
X = df.drop(columns = ['Diagnosis'])
y = df['Diagnosis']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 40, shuffle = True)


param_grids = {
    'Decision Tree': {'max_depth': [3, 5, 7, None]},
    'Random Forest': {'n_estimators': [50, 100, 200], 'max_depth': [3, 5, 7, None]},
    'K-Nearest Neighbors': {'n_neighbors': [3, 5, 7]},
    
    
}
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    
    
}


for name, model in models.items():
    grid_search = GridSearchCV(model, param_grids[name], cv = 5, scoring = 'f1')
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    report = classification_report(y_test, y_pred)
    print(f'{name} Classification Report:\n{report}\nBest Parameters: {grid_search.best_params_}\n')
```

### Lets try to improve accuracy
